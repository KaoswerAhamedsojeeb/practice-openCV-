```python
import cv2
```

```python
# capture the video using the camera
cap = cv2.VideoCapture(0) # here 0 means capturing video using the webcam and 1 means using an external camera

# get the default height and width of the camera

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#define codc and create videoweiter object 

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('videos/output_video.mp4', fourcc, 20.0, (frame_width,frame_height))


while True:
    ret, frame = cap.read()

    if not ret:
        print("could not find frame")
        break

    cv2.imshow("captured image", frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
    
```
