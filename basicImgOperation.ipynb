```python
import cv2
```

 # notes

 to load image we use : image = cv2.imread(filename, flags)

# Parameters
 filename (Required): The local relative or absolute system path to the image file.
 
 flags (Optional): Specifies the configuration color mode for reading the image.
 
# Common Read Flags
 You can pass these explicit OpenCV flag constants or use their integer equivalents:
 
 cv2.IMREAD_COLOR (or 1): Loads a color image. It strips any existing transparency / alpha channels. This is the default setting.
 cv2.IMREAD_GRAYSCALE (or 0): Converts and loads the image strictly into a 2D single-channel black-and-white format.
 cv2.IMREAD_UNCHANGED (or -1): Loads the image exactly as it is saved, safely including the alpha transparency channel if present

```python
# 1. Read the image from disk
img = cv2.imread("images/smilly_dog.png", 0)

# 2. Safety check: Ensure the file path was correct and data loaded
if img is None:
    print("Error: Could not load image. Please verify your file path.")
else:
    # 3. View the image in a GUI window
    cv2.imshow("Displayed Image Window", img)
    
    # 4. Freeze execution until any keyboard key is pressed
    cv2.waitKey(0)
    
    # 5. Clean up the system window cache 
    cv2.destroyAllWindows()
```

```python
# save the image

save = cv2.imwrite("images/BWdog.png", img)

if save:
    print("image saved successfully")
else:
    print("Error occure")
```

# finding the image dimention
 using the .shape property of the image array, which returns a tuple containing the height, width, and number of channels in that exact order

```python
img = cv2.imread("images/smilly_dog.png")

if img is not None:
   # Get dimensions (returns a tuple: Height, Width, Channels)
    height, width, channels = img.shape

print(f"Height: {height} pixels")
print(f"Width: {width} pixels")
print(f"Channels: {channels}")
```

# convert image to grayscale
 Converting images to grayscale in computer vision (CV) is done to simplify image data, significantly reducing computational complexity

```python
img= cv2.imread("images/smilly_dog.png")

if img is not None:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow("the gray image", gray_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
