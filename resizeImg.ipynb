```python
import cv2
```

```python
img = cv2.imread("images/smilly_dog.png")
```

```python
if img is None:
    print("Error: image couldn't loaded")

else:
    resized_img = cv2.resize(img, (300,300))


cv2.imshow("original image", img)
cv2.imshow("resized image", resized_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# crop an image

To crop an image in OpenCV (cv2) using Python, you use NumPy array slicing with the syntax,

cropped = image[y_start:y_end, x_start:x_end].
Because OpenCV loads images as standard NumPy arrays, there is no special function required for basic cropping.

```python
if img is None:
    print("Error: image couldn't loaded")

else:
    croped_img = img[100:500, 100:500]


cv2.imshow("original image", img)
cv2.imshow("resized image", croped_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# image rotation

Image rotation in OpenCV (CV) is performed using cv2.rotate() for exact 90° or 180° turns, 

or cv2.getRotationMatrix2D() combined with cv2.warpAffine() for arbitrary angles

```python
if img is not None:
    h, w = img.shape[:2]

    center = (h//2, w//2)

    M = cv2.getRotationMatrix2D(center, 60, 1.0)
    scaled_img = cv2.warpAffine(img,M, (h,w))

cv2.imshow("Original image is:",img)
cv2.imshow("Scaled image is:",scaled_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# drawing line and rectangle on image

```python
img3= cv2.imread("images/smilly_dog.png")

p1 = (100,100)
p2 = (300,300)
thikness = 3
color1 = [200,0,0]
color2= [0,200,0]

line_img = cv2.line(img3, p1,p2, color1, thikness)
rectan_img = cv2.rectangle(img3, p1,p2, color2, thikness)

#cv2.imshow("Line on image:",line_img)
cv2.imshow("rectangle on image",rectan_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# drawing circle on image

```python
img4= cv2.imread("images/smilly_dog.png")
center = (260,270)
circle_img = cv2.circle(img4,center, 100,[0,0,0], 4)
cv2.imshow("circle on image", img4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# adding text over image 

```python
img5= cv2.imread("images/smilly_dog.png")

# 2. Define text attributes
message = "Hello cuteee!!!"
org = [20,100]#org Bottom-left corner of the text string in the image
font = cv2.FONT_HERSHEY_SIMPLEX
scale = .8
color = (0, 0, 0)   # Green in BGR
thickness = 2

text_img = cv2.putText(img5,message,org, font, scale,color,thickness )

cv2.imshow("circle on image", img5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
